```python
#function to plot the line:

def plot_graph(m, c, x, y):
    p.scatter(x,y,  marker='x', color= 'r')
    p.xlim(0, 24)
    p.ylim(0, 30)
    ax = p.gca()
    ax.set_aspect('equal', adjustable='box')
    p.plot(x, [m*i+c for i in x])

# function to calculate error:
def cal_err (m, c, y):
    yactual= y
    ypredicted = [m*i+c for i in x]
    return sum([(i-j)**2 for i,j in zip(yactual,ypredicted)])
```
